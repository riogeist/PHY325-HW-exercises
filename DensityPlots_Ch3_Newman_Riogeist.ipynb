# Example 3.1 in Newman (Wave Interference)

```python
import math
import numpy as np
import pylab
import matplotlib.pyplot as plt
```

```python
wavelength = 5.0
k = 2*math.pi/wavelength
xi0 = 1.0
separation = 20.0 
side = 100.0 
points = 500 
spacing = side/points 
```

```python
x1 = side/2 + separation/2
y1 = side/2
x2 = side/2 - separation/2
y2 = side/2
```

```python
xi = np.empty([points,points],float)

for i in range(points):
    y = spacing*i
    for j in range(points):
        x = spacing*j
        r1 = math.sqrt((x-x1)**2+(y-y1)**2)
        r2 = math.sqrt((x-x2)**2+(y-y2)**2)
        xi[i,j] = xi0*math.sin(k*r1) + xi0*math.sin(k*r2)
```

```python
fig, ax = plt.subplots()

ax.imshow(xi,origin="lower",extent=[0,side,0,side])

plt.gray()
plt.show()
```
